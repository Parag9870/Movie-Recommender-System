```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.metrics.pairwise import cosine_similarity
from statistics import mean
from time import time, ctime
from math import ceil
```

```python
def conversion_sparse(rec, d1=None, d2=None, boolean=False):
    data = []
    i = []
    j = []
    
    for x in rec:
        
        if d1==None:
            i.append(x[0])
        else:
            i.append(d1[x[0]])
        if d2==None:
            j.append(x[1])
        else:
            j.append(d2[x[1]])
        
        if boolean==False:
            data.append(x[2])
        else:
            data.append(1)
    return sp.coo_matrix((data, (i,j)), shape = [np.amax(i)+1, np.amax(j)+1]).tocsr()
```

```python
arr_train = conversion_sparse(arr_train, arr_user, movies)
arr_genre = conversion_sparse(arr_genre, movies, arr_genres, boolean=True)
arr_mov_tags = conversion_sparse(arr_mov_tags, movies, None)
arr_actor = conversion_sparse(arr_actor, movies, actors)
arr_direct = conversion_sparse(arr_direct, movies, directors, boolean=True)
tfidf = TfidfTransformer()
arr_mov_tags = tfidf.fit_transform(arr_mov_tags)
```

```python
def item_based_filtering(train_sparse, arr_test, tags_sparse, k=175):
    test_y = []
    cos_sim_user = cosine_similarity(train_sparse, train_sparse,dense_output=False)   
    cos_sim_movie = cosine_similarity(train_sparse.transpose(), train_sparse.transpose(), dense_output=False)
    cos_sim_tag = cosine_similarity(tags_sparse, tags_sparse, dense_output=False)
    
    for index, line in enumerate(arr_test):
        user = arr_user[line[0]]
        movie = movies[line[1]]
        
        neighbor_user_ind = cos_sim_user[user].todense()
        neighbor_user_ind = np.array(neighbor_user_ind).flatten()
        neighbor_user_ind = neighbor_user_ind.argpartition(-k-1)[-k-1:]
        neighbor_user_ind = neighbor_user_ind[np.where(neighbor_user_ind != user)]
        
        user_ratings = []
        for userID in neighbor_user_ind:
            user_ratings.append(train_sparse[userID, movie])
            
        user_ratings = np.array(user_ratings)
        user_ratings = user_ratings[np.nonzero(user_ratings)]
        
        if user_ratings.size == 0:
            
            try:            
                user_rating = user_ratings.mean()
            except:
                
                user_rating = 0
        else:
            user_rating = user_ratings.mean()
            
            
        neighbor_mov_ind = cos_sim_movie[movie].todense()
        neighbor_mov_ind = np.array(neighbor_mov_ind).flatten()
        neighbor_mov_ind = neighbor_mov_ind.argpartition(-k-1)[-k-1:]
        neighbor_mov_ind = neighbor_mov_ind[np.where(neighbor_mov_ind != movie)]
        movie_ratings = []
        
        for movieID in neighbor_mov_ind:
            movie_ratings.append(train_sparse[user, movieID])
            
        movie_ratings = np.array(movie_ratings)
        movie_ratings = movie_ratings[np.nonzero(movie_ratings)]
        
        
        if movie_ratings.size == 0:
            movie_ratings = train_sparse[:, movie].data 
            movie_rating = movie_ratings.mean()
            
            if (np.isnan(movie_rating)): 
                movie_rating = 0
        else:
            movie_rating = movie_ratings.mean()
            
            
        tag_neighbor_indices = cos_sim_tag[movie].todense()
        tag_neighbor_indices = np.array(tag_neighbor_indices).flatten()
        tag_neighbor_indices = tag_neighbor_indices.argpartition(-k-1)[-k-1:]
        tag_neighbor_indices = tag_neighbor_indices[np.where(tag_neighbor_indices != movie)]
        
        new_ratings = train_sparse[:, tag_neighbor_indices]
        tag_ratings = np.true_divide(new_ratings.sum(0),(new_ratings!=0).sum(0))
        
        tag_rating = np.nanmean(tag_ratings)
        
        ratings = np.array([user_rating, movie_rating, tag_rating])
        
        ratings = ratings[np.nonzero(ratings)]
 
       
        if ratings.size == 0:
            new_rating = 3.0
        else:
            new_rating = np.mean(ratings)
            
        
        test_y.append(new_rating)

        
        if np.isnan(new_rating):
            print(index)
            print(ratings)
        
        
        if(index % 1000 == 0):
            print(ratings)
            print(index)
            print(ctime())
        
    return np.array(test_y)
```

```python
submission = item_based_filtering(arr_train, arr_test, arr_mov_tags, k=175)
```

```python
np.savetxt('out4.dat', submission, fmt='%s')
```
